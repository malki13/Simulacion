#### Camas disponibles y Costo de Atencion. 
El costo de tratar a un paciente con el nuevo coronavirus resulta elevado para cualquier pas, y a Guatemala le pega duro por su sistema de salud debilitado. El gasto en el cuidado de un paciente grave podra superar los Q50 mil, segn clculos del Ministerio de Salud.

Como una constante que lleva aos, la capacidad de los hospitales es insuficiente para alcanzar a la poblacin que requiere atencin. Hay 51 hospitales dentro del sistema pblico, con 10 mil 200 camas, mientras que el Instituto Guatemalteco de Seguridad Social (IGSS) tiene 23 hospitales y 2 mil 500 camas, esto resulta poco para cubrir a una poblacin de 16 millones 300 mil habitantes, segn el censo poblacional publicado el ao pasado.

Ante la emergencia desencadenada por el nuevo coronavirus, el gobierno habilit dos hospitales con 443 camas para atender especficamente a pacientes infectados con el virus, como respuesta para sobrellevar la crisis.

De continuar diseminndose la enfermedad, la cifra de personas enfermas con covid-19 crecer y la situacin ser insostenible, de ah que se insista en las medidas de prevencin, como el distanciamiento social y el uso de mascarilla.

Se estima que de los casos que se detectan con el virus un 20% podra llegar a requerir hospitalizacin, mientras que entre el 5% y 10% necesitar cuidados intensivos, con soporte ventilatorio.

Atender a estas personas representa un costo alto para cualquier Sistema de Salud. El Ministerio estableci en febrero pasado protocolos con el fin de organizar los servicios hospitalarios para responder la emergencia por el nuevo coronavirus, en el documento Procedimientos para la preparacin y respuesta frente al nuevo covid-19 hizo un estimado del costo en el cuidado de una persona infectada al momento de ser internada en un hospital.

Un paciente no grave requiere alrededor de Q2 mil 800 en medicamentos, entre los que se mencionan antipirticos, usados para controlar la fiebre; corticoides, que tienen efecto antiinflamatorio; antihistamnicos, los cuales se utilizan para controlar los procesos alrgicos, antibiticos de primera lnea y rehidratacin oral.

Mientras que un caso complicado que requiere ventilacin, el costo de medicamentos puede sobrepasar los Q50 mil, pues requiere antibiticos de segunda y tercera lnea -ms fuertes-, as como sedantes, entre otros frmacos.
