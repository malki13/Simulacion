Se procede a gener numeros pseudoaleatorios tomando como semilla los valores de los componentes de la computadora como son la temperatura, cpu y la RAM.  

Para poder obtener estos valores procedemos a utilizar dos librerias y definimos dos metodos para relizar los calculos necesarios para poder generar los numeros.

```python
import wmi #
import cpuinfo
import psutil
```

```python
#Obtener temperatura
w = wmi.WMI(namespace="root\wmi")
print(w.MSAcpi_ThermalZoneTemperature()[0])
temperatura_pc = w.MSAcpi_ThermalZoneTemperature()[0].CurrentTemperature
#obtener numero de calculos flops
OxC = 16 #operaciones de punto flotante por segundo del procesador Intel-i7_8750H
procesadores = float(cpuinfo.get_cpu_info()['count'])
velocidad = float(cpuinfo.get_cpu_info()['hz_advertised_friendly'].split(" ")[0])
flops = procesadores * velocidad * OxC
numero_flops=int(flops * 1024)
print(numero_flops)
#obtener uso de procesador y RAM
uso_procesador=psutil.cpu_percent()
ram=psutil.virtual_memory().total
uso_ram=psutil.virtual_memory().used
print(uso_procesador)

def cortes(num_digitos):
    digstos1 =0
    digitos2=0
    if num_digitos%2 !=0:
        digstos1 = int(num_digitos / 2)
        digitos2 = int(num_digitos / 2) + 1
    else:
        digstos1 = int(num_digitos / 2)
        digitos2 = int(num_digitos / 2)
    return digstos1,digitos2

def calculo_num(num_iteras, valor_inicio, num_digitos):
    semilla_xi = int(valor_inicio)
    aux = cortes(num_digitos)
    print("*#-ITERACIÓN*", "*Xn*", "-Xn*Xn-", "-Longitud-","-Ui-","-Rn-")
    for i in range(num_iteras):
        xn2= semilla_xi ** 2
        lon = len(str(xn2))
        ui = str(xn2)[int(lon/2)-aux[0]:int(lon/2)+aux[1]]
        rn = int(ui) / 10 ** num_digitos
        print(i, " ", semilla_xi, " ", xn2, " ", lon, " ", ui, " ", rn)
        semilla_xi=int(ui)
    print("   ")
    
```

```python
valores_iniciales =[temperatura_pc, numero_flops, uso_procesador, ram, uso_ram, 3242, 432537, 8265, 261119, 20127]
iteraciones = 16
digitos_iniciales = 4
for i in valores_iniciales:
    print("i: ", i)
    calculo_num(iteraciones, i, digitos_iniciales)
    print("  ")
```
