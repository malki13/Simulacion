# Regresion Lineal con Datos Covid-19 Ecuador



```python
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
from sklearn.metrics import mean_squared_error
from scipy.optimize import curve_fit
from scipy.optimize import fsolve
from sklearn import linear_model
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
import numpy as np #Librería numérica
import matplotlib.pyplot as plt  
from sklearn import linear_model #Regresión Lineal con scikit-learn 

url = 'covidDatos.csv'
df = pd.read_csv(url).fillna(0) 
df = df[df['location'].isin(['Ecuador'])]
df = df.loc[:,['date','total_cases', 'total_deaths']]
FMT = '%Y-%m-%d'
date = df['date']
df['date'] = date.map(lambda x : (datetime.strptime(x, FMT) - datetime.strptime("2019-12-31", FMT)).days)
```

```python
x = np.array(df.values[:,0])
y = np.array(df.values[:,1])   
z = np.array(df.values[:,2]) 


plt.scatter(x,y,label='Casos en Ecuador ', color='blue') 
plt.scatter(x,z,label='Muertos en Ecuador ', color='black')
plt.grid(True)
plt.legend()
plt.title('# de Casos');
```

```python
modelo = linear_model.LinearRegression()
modelo.fit(x.reshape((-1,1)), y)  
y_pred = modelo.predict(x.reshape((-1,1)))


plt.scatter(x,y,color='blue') 
plt.title("Regresion lineal numero de casos")
plt.plot(x,y_pred,color='black')  
plt.grid(True)
plt.show()   

print("ECUACION DE LA RECTA F(X) = MX + B")
print('M = ' + str(modelo.coef_) + ', b = ' + str(modelo.intercept_))

numero = 350
nuevo_x = np.array([int(numero)])
prediccion = modelo.predict(nuevo_x.reshape(-1,1))
print("LA PREDICCION ES IGUAL A: " , prediccion[0]) 
```

```python
modelo = linear_model.LinearRegression()
modelo.fit(x.reshape((-1,1)), z)  
y_pred = modelo.predict(x.reshape((-1,1)))


plt.title("Regresion lineal numero de muertes")
plt.scatter(x,z,color='blue')
plt.plot(x,y_pred,color='black')  
plt.grid(True)
plt.show()  

print("ECUACION DE LA RECTA F(X) = MX + B")
print('M = ' + str(modelo.coef_) + ', b = ' + str(modelo.intercept_))


numero = 359
nuevo_x = np.array([int(numero)])
prediccion = modelo.predict(nuevo_x.reshape(-1,1))
print("LA PREDICCION ES IGUAL A: " , prediccion[0]) 
```

# Regresion Logistica

```python
from scipy.optimize import curve_fit
from sklearn.linear_model import LogisticRegression  
```

```python
def modelo_logistico(x,a,b):
    return a+b*np.log(x)

x=np.arange(1,len(df)+1,1) # arreglo de x para simular el numero de dias 
y=np.array(df.values[:,1]) # casos
y1=np.array(df.values[:,2]) # muertes

exp_fit = curve_fit(modelo_logistico,x,y) #valores de casos
exp_fit1 = curve_fit(modelo_logistico,x,y1) #valores de muertes
```

```python
pred_x = list(range(min(x),max(x)+50)) # Predecir 50 dias mas
plt.rcParams['figure.figsize'] = [7, 7]
plt.rc('font', size=7)
# Real data
plt.scatter(x,y,label="Datos Reales Casos",color="red")
# Predicted exponential curve
plt.plot(pred_x, [modelo_logistico(i,exp_fit[0][0],exp_fit[0][1]) for i in pred_x], label="Modelo Logistico",color="black")
plt.legend()
plt.xlabel("Desde el 1 Marzo 2020")
plt.ylabel("Total de personas infectadas")
plt.ylim(0,max(y)*2) # Definir los limites de Y
plt.show()

```

```python
pred_x = list(range(min(x),max(x)+50)) # Predecir 50 dias mas
plt.rcParams['figure.figsize'] = [17, 17]
plt.rc('font', size=17)
# Real data
plt.scatter(x,y1,label="Datos Reales Muertes",color="red")
# Predicted exponential curve
plt.plot(pred_x, [modelo_logistico(i,exp_fit1[0][0],exp_fit1[0][1]) for i in pred_x], label="Modelo Logistico",color="black" )
plt.legend()
plt.xlabel("Desde el 1 Marzo 2020")
plt.ylabel("Total de personas infectadas")
plt.ylim(0,max(y)*2) # Definir los limites de Y
plt.show()

```
