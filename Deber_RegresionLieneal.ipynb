# Deber de Regresion Lienal


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #Regresión Lineal

        
def f(x): 
    np.random.seed(42) # para poder reproducirlo
    y = 0.1*x + 1.25 + 0.2*np.random.randn(x.shape[0]) #calculo de valores en base a una funcion
    return y

x = np.arange(0, 20, 0.5) 
y = f(x)

plt.scatter(x,y,label='data', color='blue')
plt.grid(True)
plt.title('Datos Generados');

```

```python
regresion_lineal = LinearRegression() # Realizamos la regresion lineal de sklearn
regresion_lineal.fit(x.reshape(-1,1), y) 
print("ECUACION DE LA RECTA F(X) = MX + B")
print('M = ' + str(regresion_lineal.coef_) + ', b = ' + str(regresion_lineal.intercept_))
```

```python
funcion_lineal=lambda x: regresion_lineal.coef_*x+regresion_lineal.intercept_
li=np.arange(min(x)-5.0,max(x)+5.0,0.5)
plt.scatter(x,y,label='data', color='blue')
plt.plot(li,funcion_lineal(li))
plt.grid(True)
plt.show()
```

```python
#Realizamos una predicin con predict de sklearn
nuevo_x = np.array([10]) 
prediccion = regresion_lineal.predict(nuevo_x.reshape(-1,1))
print(prediccion)
```
